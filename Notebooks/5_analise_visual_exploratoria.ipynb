## ⚙️  Categorias de risco
Traduz os números em palavras, útil em relatórios e painéis.


```python
def categorizar(valor, col):
    if col == 'fumaca':
        if valor > 1000:
            return 'Crítico'
        elif valor > 500:
            return 'Alto'
        elif valor > 200:
            return 'Moderado'
        else:
            return 'Baixo'
    if col == 'risco':
        if valor > 70:
            return 'Extremo'
        elif valor > 40:
            return 'Alto'
        elif valor > 20:
            return 'Moderado'
        else:
            return 'Baixo'

df_analise_ex_incendios['class_fumaca'] = df_analise_ex_incendios['fumaca_media'].apply(lambda x: categorizar(x, 'fumaca'))
df_analise_ex_incendios['class_risco'] = df_analise_ex_incendios['risco_fogo'].apply(lambda x: categorizar(x, 'risco'))
```
