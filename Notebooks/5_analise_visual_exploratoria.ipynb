## Corrigir a numera√ß√£o de dias sem chuva com uma m√©dia ponderada para reduzir a quantidade de casas decimais desnecess√°rias

```python
import numpy as np

# Define o limite razo√°vel de 365 dias
limite_dias_sem_chuva = 90

# Transforma valores acima disso (ou negativos, se houver) em NaN
df_analise_ex_incendios['dias_sem_chuva'] = df_analise_ex_incendios['dias_sem_chuva'].apply(lambda x: x if 0 <= x <= limite_dias_sem_chuva else np.nan)
```

## 6. Agrupamento em Faixas de Risco üå°Ô∏è

Transformar dias_sem_chuva em faixas de risco, o que facilita an√°lises comparativas e modelagens futuras.
**PS:** Adicionei mais uma coluna ao dataset aqui

```python
def classificar_risco_chuva(dias):
    if dias <= 5:
        return 'Baixo'
    elif dias <= 15:
        return 'Moderado'
    else:
        return 'Alto'

df_analise_ex_incendios['risco_dias_sem_chuva'] = df_analise_ex_incendios['dias_sem_chuva'].apply(classificar_risco_chuva)
```

```python
df_analise_ex_incendios['risco_dias_sem_chuva'].value_counts().plot(kind='bar', color='orange')
plt.title('Distribui√ß√£o das Faixas de Risco (Dias Sem Chuva)')
plt.xlabel('N√≠vel de Risco')
plt.ylabel('Frequ√™ncia')
plt.grid(True)
plt.show()
```

## Agregando os dados por munic√≠pio

```python
# Agrupar e calcular as m√©dias por munic√≠pio
df_agg_municipios = df_analise_ex_incendios.groupby('municipio', as_index=False).agg({
    'lat': 'mean',
    'lon': 'mean',
    'risco_fogo': 'mean',
    'temp_media': 'mean',
    'umidade_relativa': 'mean',
    'precipitacao': 'mean',
    'dias_sem_chuva': 'mean',
    'fumaca_media': 'mean',
    'dias_sem_chuva': 'mean'
})
```

# Juntar dados tabulares com geodados

### Geodataframe final

```python
import geopandas as gpd
import pandas as pd

# 1. Carregar o GeoJSON dos munic√≠pios de RO
gdf_municipios = gpd.read_file('../assets/jsons/municipios_ro.geojson.json')

# 2. Verificar e normalizar os nomes dos munic√≠pios (isso √© MUITO importante)
gdf_municipios["municipio"] = gdf_municipios["name"].str.strip().str.upper()
df_agg_municipios["municipio"] = df_agg_municipios["municipio"].str.strip().str.upper()

# 3. Fazer o merge (jun√ß√£o) pelo nome do munic√≠pio
gdf_merged = gdf_municipios.merge(df_agg_municipios, on="municipio", how="left")

# 4. Verificar se a jun√ß√£o deu certo
print(gdf_merged.head())
print("Total de munic√≠pios com dados:", gdf_merged['risco_fogo'].notna().sum())
```
