## ‚öôÔ∏è  Categorias de risco
Traduz os n√∫meros em palavras, √∫til em relat√≥rios e pain√©is.


```python
def categorizar(valor, col):
    if col == 'fumaca':
        if valor > 1000:
            return 'Cr√≠tico'
        elif valor > 500:
            return 'Alto'
        elif valor > 200:
            return 'Moderado'
        else:
            return 'Baixo'
    if col == 'risco':
        if valor > 70:
            return 'Extremo'
        elif valor > 40:
            return 'Alto'
        elif valor > 20:
            return 'Moderado'
        else:
            return 'Baixo'

df_analise_ex_incendios['class_fumaca'] = df_analise_ex_incendios['fumaca_media'].apply(lambda x: categorizar(x, 'fumaca'))
df_analise_ex_incendios['class_risco'] = df_analise_ex_incendios['risco_fogo'].apply(lambda x: categorizar(x, 'risco'))
```

## Verificando munic√≠pios e inconsist√™ncias para Normaliza√ß√£o

```python
import unicodedata

def normalizar(nome):
    if pd.isnull(nome):
        return ''
    nome = nome.lower()
    nome = unicodedata.normalize('NFD', nome).encode('ascii', 'ignore').decode('utf-8')
    return nome.strip()

gdf_municipios['municipio_normalizado'] = gdf_municipios['municipio'].apply(normalizar)
df_analise_ex_incendios['municipio_normalizado'] = df_analise_ex_incendios['municipio'].apply(normalizar)
```

```python
import unicodedata

def normalizar(nome):
    if pd.isnull(nome):
        return ''
    nome = nome.lower()
    nome = unicodedata.normalize('NFD', nome).encode('ascii', 'ignore').decode('utf-8')
    return nome.strip()

df_analise_ex_incendios['municipio_normalizado'] = df_analise_ex_incendios['municipio'].apply(normalizar)
gdf_municipios['municipio_normalizado'] = gdf_municipios['municipio'].apply(normalizar)
```

```python
# Filtrar gdf para conter s√≥ os munic√≠pios que existem no df de an√°lise
municipios_validos = set(df_analise_ex_incendios['municipio_normalizado'])
gdf_municipios = gdf_municipios[gdf_municipios['municipio_normalizado'].isin(municipios_validos)]
```
